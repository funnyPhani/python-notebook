```python
import requests
from bs4 import BeautifulSoup
page=int(input('enter the page num :'))
for z in range(0,page+1):
    url='https://www.indiatoday.in/science?page={}'.format(z)

    req  = requests.get(url).content
    soup = BeautifulSoup(req,'html.parser')

    data = soup.findAll('div',{'class':'detail'})

    for i in data:
        for j in i.findAll('h2'):
            print('Title :',j.text)
            for k in j.findAll('a'):
                print('Url :','https://www.indiatoday.in'+k.get('href'))
                aurl='https://www.indiatoday.in'+k.get('href')

        for l in i.findAll('p'):
            print("Description :",l.text.strip().replace('\n',""),end='')
            print()
            print('-'*15)
        req1=requests.get(aurl).content
        soup1=BeautifulSoup(req1,'html.parser')
        des=soup1.findAll('div',{'class':'description'})
        for m in des:
            print(m.text.strip().replace('\n',""),end='')
        print()
        print()
        print('-'*50)
        print()
print('data colected sucessfully......................')
```

```python
import numpy as np
def dowmloadimages(url):
    urllib.request.urlretrieve(url,r"images/{}.jpg".format(np.random.randint(1,1000000)))
```

```python
#url='https://www.indiatoday.in/science/story/how-do-birds-migrate-such-long-distances-scientists-decode-the-enigma-1818815-2021-06-24'
url=input('enter the url :')
req=requests.get(url).content
soup=BeautifulSoup(req,'html.parser')
images=soup.findAll('div',{'class':'stryimg'})
for i in images:
    for j in i.findAll('img'):
        print(j.get('data-src'))
        imagesdataurl=j.get('data-src')
        dowmloadimages(imagesdataurl)
images1=soup.findAll('div',{'class':'itgimage'})
for i in images1:
    for j in i.findAll('img'):
        print(j.get('data-src'))
        imagesdataurl=j.get('data-src')
        dowmloadimages(imagesdataurl)
```

```python
def rev(n):
    a=str(n)
    b=a[::-1]
    return b
def palindrome(n):
    a=str(n)
    b=rev(a)
    return a==b
i=s=0
while True and i<=3:
    try:
        i+=1
        print('-'*25)
        print(f"iteration :{i}")
        print('-'*25)
        print()
        no=num=int(input('enter a num :'))
        if num>0:
            if num!=int(rev(num)):
                while not palindrome(num):
                    a=rev(num)
                    s+=1
                    print("step ",s,":",num ,'+',a,"=",end=' ')
                    num=num+int(a)
                    print(num)
                break
            else:
                print(f"{num} is already palindrome")
        else:
            print('plz enter +ve num ')
            continue
    except Exception as e:
        print('error occured :',e)
print(f"{s} steps taken to get the given num {no} as palindrome ")   
```
